### Cleaning Data 

```python
#there is no null and duplicates and also datatype are accurate i just need to drop last two columns as i don't need it 
Customer_CC.drop(Customer_CC.columns[[21,22]],axis=1 ,inplace=True)
```

### Exploratory Data Analysis

```python
Numeric_Val = Customer_CC.select_dtypes(include=["number"])

corr=Numeric_Val.corr()

fig , ax = plt.subplots(figsize=(30,30))
sns.heatmap(corr , annot= True ,ax=ax)
plt.savefig("Figures/Relatioships.png" , bbox_inches="tight" )
plt.show()
```

```python
fig , ax = plt.subplots(nrows = 3 , ncols = 2 , figsize=(20,10))
ax[0][0].hist(Customer_CC['Customer_Age'])
ax[0][0].set(title = 'Customer Age Distribution')

ax[0][1].hist(Customer_CC['Months_on_book'])
ax[0][1].set(title = 'Months on book Distribution')

ax[1][0].hist(Customer_CC['Credit_Limit'])
ax[1][0].set(title = 'Credit Limit Distribution')

ax[1][1].hist(Customer_CC['Total_Trans_Amt'])
ax[1][1].set(title = 'Total Transactions Amount Distribution')

ax[2][0].hist(Customer_CC['Total_Trans_Ct'])
ax[2][0].set(title = 'Total Transactions Count')
plt.savefig("Figures/distributions.png" ,bbox_inches="tight")

plt.show()
```

```python
fig , ax = plt.subplots(nrows = 3 , ncols = 2 , figsize=(10,10))

gender_counts = Customer_CC['Gender'].value_counts()
ax[0][0].pie(gender_counts , labels = gender_counts.index , autopct = '%1.1f%%' )
ax[0][0].set(title = "Credit card customers by Gender"  )

education_counts = Customer_CC['Education_Level'].value_counts()
ax[0][1].pie(education_counts , labels = education_counts.index , autopct = '%1.1f%%')
ax[0][1].set(title = "Credit card customers by Eduication level")

martial_status_counts = Customer_CC['Marital_Status'].value_counts()
ax[1][0].pie(martial_status_counts , labels = martial_status_counts.index , autopct = '%1.1f%%')
ax[1][0].set(title = "Credit card customers by Martial Status")

card_category_count = Customer_CC['Card_Category'].value_counts()
ax[1][1].pie(card_category_count , labels=card_category_count.index , autopct = '%1.1f%%')
ax[1][1].set(title = "Credit card customers by Card Category")

attrition_counts = Customer_CC['Attrition_Flag'].value_counts()
ax[2][0].pie(attrition_counts , labels = attrition_counts.index , autopct = '%1.1f%%')
ax[2][0].set(title="Credit card customers by Attrition Flag")
ax[2][1].axis('off')
plt.savefig("Figures/PieCharts.png" ,bbox_inches="tight")

plt.show()
```

```python
fig , ax = plt.subplots(nrows = 3 , ncols = 2 , figsize = (10,10))
sns.set_style("dark")
order = Customer_CC['Income_Category'].value_counts().index
sns.countplot(data = Customer_CC , y = 'Income_Category' , palette="OrRd" , order = order, ax = ax[0][0])
ax[0][0].set(title = "Credit card customers by Income Category" , xlabel = "" , ylabel = "" )

order = Customer_CC['Dependent_count'].value_counts().index
sns.countplot(data = Customer_CC , y = 'Dependent_count' , palette="OrRd" , order = order, ax = ax[0][1])
ax[0][1].set(title = "Credit card customers by Dependent count" , xlabel = "" , ylabel = "" )

order = Customer_CC['Total_Relationship_Count'].value_counts().index
sns.countplot(data = Customer_CC , y = 'Total_Relationship_Count' , palette="OrRd" , order = order, ax = ax[1][0])
ax[1][0].set(title = "Credit card customers by  Total Relationship Count" , xlabel = "" , ylabel = "" )

order = Customer_CC['Months_Inactive_12_mon'].value_counts().index
sns.countplot(data = Customer_CC , y = 'Months_Inactive_12_mon' , palette="OrRd" , order = order, ax = ax[1][1])
ax[1][1].set(title = "Credit card customers by  Total Months Inactive" , xlabel = "" , ylabel = "" )

order = Customer_CC['Contacts_Count_12_mon'].value_counts().index
sns.countplot(data = Customer_CC , y = 'Contacts_Count_12_mon' , palette="OrRd" , order = order, ax = ax[2][0])
ax[2][0].set(title = "Credit card customers by  Total Contacts Count" , xlabel = "" , ylabel = "" )
ax[2][1].axis('off')
plt.tight_layout()
plt.savefig("Figures/barcharts.png" ,bbox_inches="tight")

plt.show()
```

### Business analysis and insights

```python
#Distribution of Attrition
fig , ax = plt.subplots(figsize=(10,10))
sns.countplot(data = Customer_CC , x = 'Attrition_Flag' , palette = 'OrRd',ax=ax)
ax.set(xlabel='Attrition Flag ' ,ylabel='Number of customers')
plt.savefig("Figures/attritions_flag.png" ,bbox_inches="tight")
```

```python
Attrited_Customer = Customer_CC[Customer_CC['Attrition_Flag']=='Attrited Customer']
```

```python
fig , ax = plt.subplots(nrows = 2 ,ncols = 2 , figsize = (20,20))
sns.countplot(data=Attrited_Customer , x='Gender' , palette = 'OrRd' , ax = ax[0][0])
ax[0][0].set(title = "Attrited Customers by Gender" , xlabel = "" , ylabel = "" )

sns.countplot(data = Attrited_Customer , y = 'Education_Level' , palette = 'OrRd' , ax = ax[0][1] )
ax[0][1].set(title = "Attrited Customers by Education Level" , xlabel = "" , ylabel = "" )

sns.countplot(data = Attrited_Customer , y = 'Marital_Status' , palette = 'OrRd' , ax = ax[1][0] )
ax[1][0].set(title = "Marital_Status" , xlabel = "" , ylabel = "" )

sns.countplot(data = Attrited_Customer , y = 'Income_Category' , palette = 'OrRd' , ax = ax[1][1] )
ax[1][1].set(title = "Income_Category" , xlabel = "" , ylabel = "" )
plt.savefig("Figures/countsplot.png" ,bbox_inches="tight")
```

```python
#Age Distribution of attritted Customers
fig , ax = plt.subplots(figsize=(20,20))
sns.histplot(data = Attrited_Customer , x = 'Customer_Age' , ax = ax ,bins = 20 )
ax.set(xlabel='Age' ,ylabel='Number of customers')
plt.savefig("Figures/agedistfor attr.png" ,bbox_inches="tight")
```

```python
#violin plots 
fig , ax = plt.subplots(nrows = 2 , ncols = 2 , figsize = (20,10))
sns.violinplot(data = Attrited_Customer , x =  'Months_Inactive_12_mon' , ax = ax[0][0])
ax[0][0].set(title = 'number of attrited customers by inactive months')

sns.violinplot(data = Attrited_Customer , x =  'Contacts_Count_12_mon' , ax = ax[0][1])
ax[0][1].set(title = 'number of attrited customers by Contacts Counts')

sns.violinplot(data = Attrited_Customer , x =  'Total_Trans_Ct' , ax = ax[1][0])
ax[1][0].set(title = 'number of attrited customers by transactions Counts')
ax[1][1].axis('off')
plt.savefig("Figures/voilinplots.png" ,bbox_inches="tight")

```

```python
fig , ax = plt.subplots(nrows = 2 , ncols = 2 , figsize = (20,10))
sns.boxplot(data = Attrited_Customer , x =  'Credit_Limit' , ax = ax[0][0])
ax[0][0].set(title = 'number of attrited customers by Credit Limit')

sns.boxplot(data = Attrited_Customer , x =  'Total_Trans_Amt' , ax = ax[0][1])
ax[0][1].set(title = 'number of attrited customers by Transactions Amount')

sns.boxplot(data = Attrited_Customer , x =  'Avg_Utilization_Ratio' , ax = ax[1][0])
ax[1][0].set(title = 'number of attrited customers by Avg Utilization Ratio')
ax[1][1].axis('off')
plt.savefig("Figures/boxplots.png" ,bbox_inches="tight")
```
